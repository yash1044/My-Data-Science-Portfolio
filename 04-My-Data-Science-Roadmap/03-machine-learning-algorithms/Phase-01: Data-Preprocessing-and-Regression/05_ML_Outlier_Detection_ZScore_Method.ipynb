```python
import pandas as pd
```

```python
l = [29,10,45,46,68,34,65,89,400,92,77,66,69,65,80,77,21,37,42]
```

```python
data = pd.DataFrame(l,columns=["value"])
data
```

```python
data["value"].mean()
```

```python
data["value"].std()
```

```python
z_score = (data["value"] - data["value"].mean())/data["value"].std()
```

```python
z_score
```

```python
data["z_score"] = z_score
```

```python
data
```

```python
data.hist()
```

```python
data[(data["z_score"]>3) | (data["z_score"]<-3)]
```

```python
val = data[(data["z_score"]<3) & (data["z_score"]>-3)]
val
```

```python
val.hist()
```
