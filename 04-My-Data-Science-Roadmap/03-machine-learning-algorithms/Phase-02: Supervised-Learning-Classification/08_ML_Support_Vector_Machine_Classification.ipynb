```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\ML\ML-6 Logistics Regression Files16958214402 (1).txt")
data
```

```python
x = data.drop("bought_insurance",axis=1)
y = data["bought_insurance"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train) * 100
```

```python
model.score(x_test,y_test) * 100
```

```python
pri = model.predict(x_test)
pri
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pri)
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
precision_score(y_test,pri)
```

```python
recall_score(y_test,pri)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pri))
```
