```python
def cal(x,y):
    sum = x + y
    sub = x - y
    mul = x * y
    divi = x / y
    power = x ** y
    print(sum,',',x,y,"is the sum")
    print(sub,',',x,y,"is the sub")
    print(mul,',',x,y,"is the mul")
    print(divi,',',x,y,"is the divi")
    print(power,',',x,y,"is the power")
```

```python
cal(20,10)
```

```python
def mul_elements(*a):
    ans = 1
    for i in a:
        ans = ans * i
        print(ans)
```

```python
mul_elements(1,2,3,4,5)
```

```python
def all_sum(*a):
    ans = 0
    for i in a:
        ans = ans + i
    print(ans)
```

```python
all_sum(10,20,30)
```

```python
def factorial(a):
    fact = 1
    for i in range(1,a+1):
        fact = fact * i
    print(fact)
```

```python
factorial(6)
```
