```python
num = int(input("Enter Number"))

if num >= 100:
    print("Big Positive")
elif num >= 60:
    print("Med Positive")
elif num >= 20:
    print("Small Positive")
elif num >= 0:
    print("Very Small Positive")
else:
    print("Invalid Input")
```

```python
c = 1

while c < 20:
    if c %2 == 0:
        print(c,"even")
    else:
        print(c,"odd")
    c += 1
```
