```python
y = np.array([20,35,10,45])
l = ["ind","usa","uk","china"]
plt.pie(y,labels=l,autopct="%1.2f%%",textprops={"color":"white"})
plt.show()
```

```python
y = [20,35,10,45]
l = ["ind","usa","uk","china"]
plt.pie(y,labels=l,autopct="%1.2f%%",textprops={"color":"white"},colors=["k","g","y","r"],)
plt.show()
```

```python
y = [20,35,10,45]
l = ["ind","usa","uk","china"]
plt.pie(y,labels=l,autopct="%1.2f%%",textprops={"color":"white"},colors=["k","g","y","r"],shadow=True,explode=[0.2,0.5,0.2,0],startangle=45)
plt.show()
```
