```python
data = data.drop("UnderRisk",axis=1)
```

### Make 3 groups to better prediction and build model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
kmeans.fit(data)
```

```python
kmeans.cluster_centers_
```

```python
data["cluster"] = kmeans.labels_
```

```python
data.head()
```

### Plot dendrogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(data.drop("cluster", axis=1), method='ward')

dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.show()
```

### Model Evalution

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
labels = agg.fit_predict(data.drop("cluster", axis=1))
```

```python
data["label"] = labels
```

```python
data["label"].value_counts()
```

```python
score = silhouette_score(data.drop(["cluster","label"], axis=1), data["label"])

print("Silhouette Score:", score) # Nearly score 1 is show better clusters
```
